```
import matplotlib.pyplot as plt
import datetime
import numpy as np

import re
regex = re.compile("(오전 \d+:\d+|오후 \d+:\d+)")
# regex = re.compile("(2019.+오전 \d+:\d+|2019.+오후 \d+:\d+)")
number = re.compile("\d+")
result = regex.findall(text)

x_keys=np.array([datetime.datetime(2019,3,1,0,0)+datetime.timedelta(minutes=i) for i in range(1440)])
x=dict.fromkeys(x_keys,0)
# # week = [x]*10
# # print(week)
# idx=0
for i in result:
#     print(i)
    j=number.findall(i)
#     print(j)
#     print(datetime.datetime(int(j[0]),int(j[1]),int(j[2])).weekday())

    if(i[0:2] == '오전'):
        if(j[0]=="12"):
            j[0]=0
        x[datetime.datetime(2019,3,1,int(j[0]),int(j[1]))]=x[datetime.datetime(2019,3,1,int(j[0]),int(j[1]))]+1
    else:
        if(j[0]=="12"):
            j[0]=0
        x[datetime.datetime(2019,3,1,(int(j[0])+12),int(j[1]))]=x[datetime.datetime(2019,3,1,(int(j[0])+12),int(j[1]))]+1
    
plt.plot(x.keys(),x.values())
# plt.xticks(range(1,1440,60),range(1,1440,60))
axes = plt.gca()
axes.set_ylim([0,max(x.values())])
plt.rcParams['figure.figsize'] =[30,5]
plt.show()
```
